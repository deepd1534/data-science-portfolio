```
import numpy as np,pandas as pd,matplotlib.pyplot as plt,seaborn as sns
```

```
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```
df.head()
```

```
print(df.shape)
print(df['species'].value_counts())
```

## UNIVARIATE ANALYSIS

- Simplest type of analysis done with only one feature
- Data points can be classified with one feature but one feature is never enough most of the times.

```
#DIVIDING THE DATASET BASED ON SPECIE

df_setosa = df.loc[df['species'] == 'setosa']
df_virginica = df.loc[df['species'] == 'virginica']
df_versicolor = df.loc[df['species'] == 'versicolor']
```

```
#UNIVARIATE DIAGRAM OF PETAL LENGTH OF VARIOUS SPECIES OF IRIS DATASET

sns.set_style("whitegrid")
plt.plot(df_setosa['petal_length'],np.zeros_like(df_setosa['petal_length']),'x')
plt.plot(df_virginica['petal_length'],np.zeros_like(df_virginica['petal_length']),'x')
plt.plot(df_versicolor['petal_length'],np.zeros_like(df_versicolor['petal_length']),'x')
plt.xlabel("Petal Length")
plt.show()
```

## BIVARIATE ANALYSIS

- We consider two features for bivariate analysis, Useful to decide the algorithm.

```
#BIVARIATE ANALYSIS OF sepal_length AND sepal_width

sns.FacetGrid(df,hue='species',height = 5).map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

## MULTIVARIATE ANALYSIS
- Multiple features are taken and plotted for multivariate analysis.
- They give us an idea about correlation and help us in determining the algorithm.

```
#MULTIVARIATE ANALYSIS OF VARIOUS FEATURES OF IRIS DATASET

sns.pairplot(df,hue='species',height = 3)
plt.show()
```
